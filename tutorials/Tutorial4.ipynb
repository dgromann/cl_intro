## **Lesson 0.3: Numpy**

[NumPy](https://numpy.org/doc/stable/) is a Python library, which adds support for large, multi-dimensional arrays and matrices, along with a large collection of optimized, high-level mathematical functions to operate on these arrays.

````python
# Import numpy
import numpy as np
````

````python
# Create numpy arrays from lists
list1 = [1,2,3]
x = np.array(list1)

y = np.array([[3,4,5]])
z = np.array([[6,7],[8,9]])

# Let's take a look at their shapes.
# When working with numpy arrays, .shape will be a very useful debugging tool
print(x)
print("Shape of x: ", x.shape)
print()
print(y)
print("Shape of y: ", y.shape)
print()
print(z)
print("Shape of z: ", z.shape)
````

Vectors can be represented as 1-D arrays of shape (N,), e.g. `x` above, or 2-D arrays of shape (N, 1) or (1, N), e.g. `y` above. But it's important to note that the shapes (N,), (N, 1), and (1,N) are not the same and may result in different behavior, for which we will see examples later/next time.

Matrices are generally represented as 2-D arrays of shape (M, N), e.g. `z` above.

The best way to ensure your code gives you the behavior you expect is to keep track of your array shapes and try out small test cases or refer back to documentation when you are unsure.

**Gradients and Backpropagation**

If you set the flag  ```.requires_grad``` on a ```torch.Tensor``` to ```True``` the program will track all operations on it in order to enable later operations, such as backpropagation, which is very important to neural networks.

When you finish all computations on your tensor, you can then simply call the function ```.backward()``` and have all the gradients computed automatically. The gradient will then automatically be accumulated in the attribute ```.grad```.

Initially there is nothing to show if no operations have been performed on the tensor:

````python
x = torch.tensor([2.], requires_grad=True)
pp.pprint(x.grad)
````

Once operations have been performed, you can calculate the gradient, e.g. of y with respect to x in the following example:

````python
y = x * x * 3 # 3x^2
y.backward()
pp.pprint(x.grad) # d(y)/d(x) = d(3x^2)/d(x) = 6x = 12
````

If you wish to disconnect a specific tensor from this process of tracking all operations, you can call the function ```.detach()```. This prevents future computations from being tracked. You can alternatively wrap the code block in a function ```with torch.no_grad()``` which does not track the operations on any variables included in the block. This is particularly helpful if you wish to evaluate a model that has trainable parameters with *required_grad=True* flags but for which we don't need the gradients in evaluation.


````python
# Gradients
# Let's calculate and print the gradietn (d(out)/dx) and print it
x.backward()
pp.pprint(x.grad)

# Stop autograd from tracking
x.detach
pp.pprint((x ** 2).requires_grad)

with torch.no_grad():
    pp.pprint((x ** 2).requires_grad)
````
